## String Concatenation in Python

* Joining of two or more strings to make them one is called string concatenation in Python.

* In general, Python provides `(+)` operator to concatenate multiple strings. 
* We can concatenate two or more strings, or a string with a number or symbol to display information as per requirements.

#### Example: Python program to demonstrate the string concatenation.

```python
# Take first name as input from the user.
first_name = input('Enter your first name: ')

# Take middle name as input from the user.
middle_name = input('Enter your middle name: ')

# Take last name as input from the user.
last_name = input('Enter your last name: ')

# Concatenate strings using + operator.
string_concatenation = first_name + " " + middle_name + " " +last_name
print("\nFull name:",string_concatenation)
```

### Ways to Concatenate Strings in Python

String concatenation is heavily used in Python programming. 
- Using `+` operator
- Using `join()` method
- By using `format()` function

### String Concatenation using `+` Operator

- Example1:

```python
# Python program to join three strings.
str1 = 'Python'
str2 = 'Programming'
str3 = 'Language' 
concatenated_string = str1 +" "+ str2 +" "+ str3
print('Joined string:',concatenated_string)
```

- Example 2:

```python
# Python program to join a number with a string.
num = 25
string = 'twenty-five'
# Joining a number with string.
concatenated_string = num + string
print('Joined string:',concatenated_string)
```

### String Concatenation using `join()` Method

```python
# Python program to concatenate strings using join() method.
first_name = input("Enter first name:")
middle_name = input("Enter middle name:")
last_name = input("Enter last name:")

# This statement inserts blank space between each sequence of strings.
full_name = " ".join((first_name, middle_name, last_name))
print("\nFull name:",full_name)
```

```python
# Python program to insert a number between each character of string using join() method.
num = "3"
string = 'Abdul'
joined_string = num.join(string)
print(joined_string)
```

### Joining Strings using Formatting Operator `(%)`

We can also use a string formatting operator `(%)` for joining two or more strings or string and number. Let’s take a simple example based on it.

```python
# Python program to concatenate strings using formatting operator %.
first_name = 'Abdul'
last_name = 'Aziz'
full_name = "%s %s" %(first_name, last_name)
print('Full name:',full_name)

day = 21
month = 'Nov'
year = 2002
dob = "%s %s %s" %(day, month, year)
print("\nDate of birth:",dob)
```

```python
# Python program to concatenate strings using format() function.
first_name = 'Abdul'
middle_name = 'Aziz'
last_name = "Md"
full_name = "{} {} {}".format(first_name, middle_name, last_name)
print('Full name:',full_name)

day = 12
month = 'Nov'
year = 2010
dob = "{} {} {}".format(day, month, year)
print("\nDate of birth:",dob)
```

## Access Characters by Index

- You can access individual characters in a string using an index in square brackets. The string indexing starts from 0.

- You can also access a string by negative indexing. A negative string index counts from the end of the string.

- The indices for the elements in a string are illustrated as below:



```python
# Indexing

S = 'ABCDEFGHI'
print(S[0])    # Prints A
print(S[4])    # Prints E

# Negative Indexing
print(S[-1])    # Prints I
print(S[-6])    # Prints D
```

### String Slicing using slice operator

- We can obtain a substring of an original string using slice operator. 
- It is an easy and convenient way to slice a string. 
- To slice or access a range of characters in a string or create a substring, use the slice operator `[:]`.

- If `S` is a string, the expression `S [ start : stop : step ]` returns the portion of the string from `index start` to `index stop`, at a step size `step`.



#### Basic Example
Here is a basic example of string slicing.





```python
S = 'ABCDEFGHI'
print(S[2:7]) # CDEFG
```

**Slice with Negative Indices**

You can also specify negative indices while slicing a string.





```python
S = 'ABCDEFGHI'
print(S[-7:-2]) # CDEFG
```

#### Specify Step of the Slicing

- You can specify the step of the slicing using `step` parameter. The step parameter is optional and by default 1.



```python
# Return every 2nd item between position 2 to 7
S = 'ABCDEFGHI'
print(S[2:7:2]) # CEG
```

#### Slice at Beginning & End

- Omitting the `start` index starts the slice from the index 0. 
- Meaning, `S[:stop]` is equivalent to `S[0:stop]`

```python
# Slice first three characters from the string
S = 'ABCDEFGHI'
print(S[:3])    # ABC
```

- Whereas, omitting the stop index extends the slice to the end of the string. 
- Meaning, `S[start:]` is equivalent to `S[start:len(S)]`

```python
# Slice last three characters from the string
S = 'ABCDEFGHI'
print(S[6:])    # GHI
```

#### Reverse a String

You can reverse a string by omitting both start and stop indices and specifying a step as -1.

```python
S = 'ABCDEFGHI'
print(S[::-1]) # IHGFEDCBA
```

## Python String Methods

### 1. Python String Split Method

In python, the string `split()` method is useful to split the string into a list that contains substrings, and those are separated by the specified separator.



- String Split Method Syntax:

`string.split(separator, maxsplit)`

If you observe the above syntax, the `split()` method is accepting two parameters `(separator, maxsplit)` and those are optional.

- **separator -** It’s a delimiter to split the given string into substrings, and it’s optional. In case if the delimiter is not mentioned, by default, it will consider whitespace as a separator to split the strings.
- **maxsplit -** It is useful to define the maximum number of splits, and it’s optional. By default, the maxsplit value is considered as -1 that means no limit on the number of splits.

```python
msg1 = "Welcome to Pyhon Inernahip"
msg2 = "Suresh, Rohini, Raju"
msg3 = "Python$Java$C#"
lst1 = msg1.split()
lst2 = msg2.split(", ")
lst3 = msg3.split("$")
print(lst1)
print(lst2)
print(lst3)
```

```python
msg1 = "Suresh,Rohini,Raju,Lakshmi"
lst1 = msg1.split(",")
lst2 = msg1.split(",",1)
lst3 = msg1.split(",",2)
print(lst1)
print(lst2)
print(lst3)
```

### 2. String Replace Method

In python, the string replace() method is useful to replace the specified substring or character in all occurrences of the given string.



- String Replace Method Syntax

`string.replace(old string/char, new string/char, count)`

If you observe the above syntax, the `replace()` method accepts three parameters `(old string/char, new string/char, count)`, and the `count` parameter is optional.

- **old string/char -** It’s the old string/character you want to replace.
- **new string/char -** The new string/character will replace the old string/character.
- **count -** It’s optional and useful to define the number of times the old substring to replace with the new substring.

```python
msg1 = "Hi Guest Hi, Learn Python"
rmsg1 = msg1.replace("Hi","Hello")
msg2 = "aaaacc"
rmsg2 = msg2.replace("a","b")
msg3 = "1 2 3 4 5 6 7"
rmsg3 = msg3.replace(" ",",")
print(rmsg1)
print(rmsg2)
print(rmsg3)
```

```python
msg1 = "Hi Guest Hi Hi Hi Hi HI HI HI"
rmsg1 = msg1.replace("Hi","Hello")
rmsg2 = msg1.replace("Hi","Hello",0)
rmsg3 = msg1.replace("Hi","Hello",1)
rmsg4 = msg1.replace("Hi","Hello",2)
print(rmsg1)
print(rmsg2)
print(rmsg3)
print(rmsg4)
```

### 3. String Find Method

- In python, the string `find()` method is useful to search the string for specified substring and return the index position of the first occurrence of the specified substring. 
- In case if the specified substring/character is not found, it will return `-1`.



- String Find Method Syntax

`string.find(string/char, startIndex, endIndex)`

If you observe the above syntax, the `find()` method accepts three parameters `(string/char, startIndex, endIndex)`. Here, `startIndex`and `endIndex` parameters are optional.

- **string/char -** It’s the substring/character you want to search in the string.
- **startIndex -** It’s optional and indicates from which index position the search can start. By default, it's 0.
- **endIndex -** It’s optional and indicates at which index position the search can end. By default, it's the end of the string.

```python
msg = "Hi Guest hi, welcome to Python Internship"
rmsg1 = msg.find("hi")
rmsg2 = msg.find("Python")
rmsg3 = msg.find("to")
rmsg4 = msg.find("Hi")
print("Substring 'hi':", rmsg1)
print("Substring 'Python':", rmsg2)
print("Substring 'to':", rmsg3)
print("Substring 'Hi':", rmsg4)
```

```python
msg = "Hi Friends hi, welcome to Pyhon Programming"
rmsg1 = msg.find("to",4,10)
rmsg2 = msg.find("hi",1,15)
rmsg3 = msg.find("Guest", 2, 8)
print("Substring 'to':", rmsg1)
print("Substring 'hi':", rmsg2)
print("Substring 'Guest':", rmsg3)
```
