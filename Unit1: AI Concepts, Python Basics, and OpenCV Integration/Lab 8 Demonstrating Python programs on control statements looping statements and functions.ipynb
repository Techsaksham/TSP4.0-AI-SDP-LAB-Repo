#### Output Formatting

#### Formatting Output using String Modulo Operator (%)
The % operator helps us to format the string. It interprets the left argument like printf() format string applied to the right argument. Python doesn't provide the printf() function.

```python
#integer and float value  
print("integer : % 2d, float : % 5.2f" %(1, 05.333))    
#integer value  
print("Total Player : % 3d, Batsman : % 2d" %(24, 12))   
#octal value  
print("% 5.3o"% (50))   
#exponential value  
print("% 10.2E"% (456.1458))  
```

#### Formatting Output Using the Format Method
The format() method takes more manual effort than other methods. We use {} to mark the substitution of variables and provide detailed formatting directives, but we also need to provide the formatted information. 

```python
# using format() method   
print('{} {}'.format('Hello', 'World'))   
    
# using format() method and referring    
# a position of the object   
print('{0} to {1}'.format('Welcome', 'Edunet'))   
print('{1} to {0}'.format('Welcome', 'Edunet'))  
```

### if Statement

```python
number = 10

# check if number is greater than 0
if number > 0:
    print('Number is positive')

print('This statement always executes')
```

### if...else Statement

```python
number = 10

if number > 0:
    print('Positive number')

else:
    print('Negative number')

print('This statement always executes')
```

```python
number = -5

if number > 0:
    print('Positive number')

else:
    print('Negative number')

print('This statement always executes')
```

### if…elif…else Statement

```python
number = 0

if number > 0:
    print('Positive number')

elif number <0:
    print('Negative number')

else:
    print('Zero')

print('This statement is always executed')
```

### Nested if Statements

```python
number = 5

# outer if statement
if number >= 0:
    # inner if statement
    if number == 0:
      print('Number is 0')
    
    # inner else statement
    else:
        print('Number is positive')

# outer else statement
else:
    print('Number is negative')
```

### Loop Control Statements

### Break

```python
for i in range(10):
  if i == 5:
    break
  print(i)
```

### Continue

```python
for i in range(10):
  if i % 2 == 0:
    continue
  print(i)
```

### Functions

A function is a block of code which only runs when it is called.
You can pass data, known as parameters, into a function.
A function can return data as a result.


```python
def greet():
    print('Hello World!')
```

```python
greet()
```

```python
def greet():
    print('Hello World!')

# call the function
greet()

print('Outside function')
```

```python
# function with two arguments
def add_numbers(num1, num2):
    sum = num1 + num2
    print("Sum: ", sum)

# function call with two values
add_numbers(5, 4)
```

```python
# function definition
def find_square(num):
    result = num * num
    return result

# function call
square = find_square(3)

print('Square:', square)
```

### Lambda/Anonymous Function

lambda argument(s) : expression 

```python
greet = lambda : print('Hello World')
```

```python
# call the lambda
greet()
```

```python
x = lambda a, b : a * b
print(x(5, 6))
```

### call by value

```python
# Python code to demonstrate
# call by value
string = "Edunet"
 
def test(string):
    string = "Sarvesh"
    print("Inside Function:", string)
 
# Driver's code
test(string)
print("Outside Function:", string)
```

### call by reference

```python
# Python code to demonstrate
# call by reference
 
def add_more(list):
    list.append(50)
    print("Inside Function", list)
 
# Driver's code
mylist = [10, 20, 30, 40]
 
add_more(mylist)
print("Outside Function:", mylist)
```
